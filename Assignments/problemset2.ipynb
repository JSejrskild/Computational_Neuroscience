# Equilibrium of differential equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

# The example from the prerecorded lecture

```python
# Duration of stimulus in ms
T = 100

# Time step (sample rate) in ms
dt = 0.1

# Vector of sampled times
t = np.arange(0, T, dt) # from 0 to T in steps of dt

# Total number of time steps
N = t.shape[0]

# Vector of state values
x = np.zeros(N) 

# Set the initial state
x[0] = 1

# Set constants
R = 1
C = 1
Vs = 5

# Step 2 - System perturbation
xp = np.zeros(N)
xp[400:600] = 1 # We excite the system in an interval of 200 steps out of the 1000 steps

for i in range(1, N):
    dt = t[i] - t[i-1] # time step
    dxdt = (1/R*C) * (Vs - x[i-1]) - xp[i-1] # rate of change of x
    x[i] = x[i-1] + dxdt * dt # update x
    
plt.plot(t, x)
plt.xlabel('Time (ms)')
plt.ylabel('State - x(t)')
plt.title('Euhlers method approximation')
plt.show()
```

Q 19 from the problem set.
Does the following dynamical system have a stable equilibrium point and if so where is it? 

$$
\begin{align*}
\frac{d}{dt}x(t) &= b - a x^2(t) \\
a &> 0\\ 
b &> 0
\end{align*}
$$
Answer: No stable equilibrium

Then adapted for Q 20

Does the following dynamical system have a stable equilibrium point and if so where is it? 

$$
\begin{align*}
\frac{d}{dt}x(t) &= b + a x^2(t) \\
a &> 0\\ 
b &> 0
\end{align*}
$$
Answer: Stable equilibrium at 1 unless very large negative pertubations

```python
# Duration of stimulus in ms
T = 100

# Time step (sample rate) in ms
dt = 0.1

# Vector of sampled times
t = np.arange(0, T, dt) # from 0 to T in steps of dt

# Total number of time steps
N = t.shape[0]
print(N)
# Vector of state values
x = np.zeros(N) 

# Set the initial state
x[0] = 0

# Set constants
a = 1
b = 1

# Step 2 - System perturbation
xp = np.zeros(N)
xp[5:100] = 10 # We excite the system in an interval of 200 steps out of the 1000 steps

for i in range(1, N):
    #dt = t[i] - t[i-1] # time step
    dxdt = b - (a*(x[i-1]**2)) + xp[i-1] # rate of change of x
    x[i] = x[i-1] + dxdt * dt # update x
    
plt.plot(t, x)
plt.xlabel('Time (ms)')
plt.ylabel('State - x(t)')
plt.title('Euhlers method approximation')
plt.show()
```
