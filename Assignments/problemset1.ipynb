```python

# Generate x and y values
x=np.linspace(-3, 3, 100)
y=np.linspace(-3, 3, 100)
X, Y=np.meshgrid(x, y)

# Generate the 2D Gaussian distribution
Z=Z=np.exp(-(X**2 + Y**2))

# Plot the Gaussian distribution as a contour plot
plt.contour(X, Y, Z)

# Show the plot
plt.show()
```

```python

# Generate x and y values
x=np.linspace(-3, 3, 100)
y=np.linspace(-3, 3, 100)
X, Y=np.meshgrid(x, y)

# Generate the 2D Gaussian distribution
Z=np.exp(-(X**2 + Y**2))

# Plot the Gaussian distribution as a heatmap
plt.imshow(Z, cmap="hot")

# Add a colorbar
plt.colorbar()

# Show the plot
plt.show()
```
