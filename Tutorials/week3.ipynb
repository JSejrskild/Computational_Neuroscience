```python
t = np.linspace(-np.pi, np.pi, 100)
sint = np.sin(t)
cost = np.cos(t)
plt.plot(sint, label = 'f(t)=sin(t)')
plt.plot(cost, label = "f(t)=cos(t)")
plt.legend() #This ensures that the labels are displayed - super smart
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100) # from 0 to 2pi
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Creating 50 random x and y's
x = np.random.rand(1000)
y = np.random.rand(1000)

#Sizes based on the distance from the center
sizes = np.sqrt((x - 0.5)**2 + (y - 0.5)**2) * 100 

#Colors
colors = np.random.rand(1000) #Random colors

plt.scatter(x, y, s = sizes, c = colors)
plt.show()
```

```python
US = [46, 37, 38]  # gold, silver, bronze
GB = [27, 23, 17]
China = [26, 18, 26]
Russia = [19, 18, 19]
Germany = [17, 10, 15]

# Set up the labels and positions for the bars
labels = ["gold", "silver", "bronze"]
x = np.arange(len(labels))  # The label positions for each medal type

# Create the stacked bar plot
plt.bar(x, US, label='US', color='gold')
plt.bar(x, GB, bottom=US, label='GB', color='silver')
plt.bar(x, China, bottom=np.array(US)+np.array(GB), label='China', color='red')
plt.bar(x, Russia, bottom=np.array(US)+np.array(GB)+np.array(China), label='Russia', color='blue')
plt.bar(x, Germany, bottom=np.array(US)+np.array(GB)+np.array(China)+np.array(Russia), label='Germany', color='green')

# Add labels and title
plt.ylabel('Number of Medals')
plt.title('Stacked Bar Plot of Medals by Country')
plt.xticks(x, labels)  # Set the x-ticks to the medal types
plt.legend()
plt.show()
```

```python
np.random.seed(42)
data = np.random.normal(size=(5, 100))
corr = np.corrcoef(data)
fig, ax = plt.subplots()
im = ax.imshow(corr, cmap='coolwarm')
cbar = ax.figure.colorbar(im, ax=ax)
for i in range(5):
    for j in range(5):
        text = ax.text(j, i, f'{corr[i,j]:.2f}', ha='center', va='center', color='black')
ax.set_xticks(np.arange(5))
ax.set_yticks(np.arange(5))
ax.set_xticklabels(['Var 1', 'Var 2', 'Var 3', 'Var 4', 'Var 5'])
ax.set_yticklabels(['Var 1', 'Var 2', 'Var 3', 'Var 4', 'Var 5'])
plt.show()
```
